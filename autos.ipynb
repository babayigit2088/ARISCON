```python
top_brands = autos["brand"].value_counts().index[:8]
top_brands
```

We can learn the average price of the top brands.

```python
top_brands_mean_price = {} 
for brand in top_brands:
    brand_only = autos.loc[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    top_brands_mean_price[brand] = int(mean_price)
```

```python
top_brands_mean_price
```

```python
top_brand_mileage_mean={}
for brand in top_brands:
    brand_only=autos.loc[autos["brand"]==brand]
    mean_mileage=brand_only["odometer_km"].mean()
    top_brand_mileage_mean[brand]=int(mean_mileage)
    
```

```python
top_brand_mileage_mean
```

```python
mean_price_series=pd.Series(top_brands_mean_price)
```

```python
mean_price_series
```

```python
mean_mileage_series=pd.Series(top_brand_mileage_mean)
```

```python
mean_mileage_series
```

```python
data_frame=pd.DataFrame(mean_price_series, columns=["mean_price"])
data_frame
```

```python
data_frame["mean_mileage"]=mean_mileage_series
```
